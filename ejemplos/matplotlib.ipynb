```python
import matplotlib.pyplot as plt
```

```python
# Datos a graficar
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Crear el gráfico
plt.plot(x, y)

# Mostrar el gráfico en pantalla
plt.show()
```

```python
plt.plot(x, y, color='blue', linestyle='--', marker='o')
plt.title("Gráfico de ejemplo")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()

```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]
plt.scatter(x, y, color='green')
plt.title("Gráfico de dispersión")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

```python
# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Título del Gráfico")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
plt.show()
```

```python
#numpy

import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python

#pandas

import pandas as pd

data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40]
})

data.plot(kind='line', x='A', y='B')
plt.show()
```
